## Анимация детерминированной динамики модели SIR

### Подготовка окружения

Подключаем необходимые пакеты и модули.

```julia
using DrWatson
@quickactivate "project"
include(srcdir("SIRPetri.jl"))
using .SIRPetri
using Plots
```

### Параметры модели

Задаём параметры для анимации:
- `β = 0.3` — коэффициент заражения
- `γ = 0.1` — коэффициент выздоровления
- `tmax = 100.0` — максимальное время симуляции

```julia
β = 0.3
γ = 0.1
tmax = 100.0
```

### Запуск симуляции

Создаём сеть Петри и запускаем детерминированную симуляцию
с шагом сохранения 0.5 для плавности анимации.

```julia
net, u0, _ = build_sir_network(β, γ)
df = simulate_deterministic(net, u0, (0.0, tmax), saveat = 0.5, rates = [β, γ])
```

### Создание анимации

Создаём анимацию, где каждый кадр показывает столбчатую диаграмму
количества людей в каждой группе (S, I, R) в соответствующий момент времени.
Берём каждый 5-й кадр для оптимальной плавности.

```julia
anim = Animation()
for i in 1:5:length(df.time)
    p = bar(
        ["S", "I", "R"],
        [df.S[i], df.I[i], df.R[i]],
        ylims = (0, 1000),
        title = "Time = $(round(df.time[i], digits=1))",
        ylabel = "Population",
        label = "",
        color = [:green, :red, :blue]
    )
    frame(anim, p)
end
```

### Сохранение анимации

Сохраняем анимацию в GIF-файл с частотой 5 кадров в секунду.

```julia
gif(anim, plotsdir("sir_animation.gif"), fps = 5)
```
